# MOVIE ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

movies_df = pd.read_csv('data/movie.csv', sep='\t')

print(movies_df.head())
```

### Top movies by gross

```python
top_movies = movies_df.sort_values(by='Lifetime Gross', ascending=False).head(10)

top_movies['Lifetime Gross'] = top_movies['Lifetime Gross'].apply(lambda x: '${:,.2f}'.format(x))

top_movies = top_movies[[ '#1 Release', 'Lifetime Gross']]

print(top_movies)
```

### Total gross revenue by brand graph

```python
plt.figure(figsize=(12, 20))  
sns.barplot(x='Total', y='Brand', data=movies_df, estimator=sum)
plt.title('Total Gross Revenue by Brand', fontsize=16) 
plt.xlabel('Total Gross Revenue', fontsize=14)  
plt.ylabel('Brand', fontsize=14)  
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12) 
plt.show()
```

### Top Releases Analysis

```python
top_releases = movies_df.groupby('Brand')['#1 Release'].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_releases.values, y=top_releases.index, hue=top_releases.index, legend=False, palette='viridis')
plt.title('Top Releases by Brand')
plt.xlabel('Number of Top Releases')
plt.ylabel('Brand')
plt.show()
```
